```python

query = """
SELECT product_id, SUM(sale_price * quantity) as total_revenue 
FROM orders 
GROUP BY product_id
ORDER BY total_revenue DESC 
LIMIT 10
"""
df_top_revenue = pd.read_sql(query, engine)
print("Top 10 Orders by Total Revenue (Sale Price × Quantity):")
display(df_top_revenue)


```

```python
query = """
SELECT 
    region,
    SUM(quantity) as total_quantity
FROM orders
GROUP BY region
ORDER BY total_quantity DESC
LIMIT 10

"""
df_top_quantity = pd.read_sql(query, engine)
print("Top 10 quantity by region:")
display(df_top_quantity)
```

```python
query = """
WITH RankedProducts AS (
    SELECT 
        region,
        product_id,
        SUM(quantity) as total_quantity,
        ROW_NUMBER() OVER (PARTITION BY region ORDER BY SUM(quantity) DESC) as rank
    FROM orders
    GROUP BY region, product_id
)
SELECT 
    region,
    product_id,
    total_quantity
FROM RankedProducts
WHERE rank <= 5
ORDER BY region, total_quantity DESC
"""
df_top_by_region = pd.read_sql(query, engine)
print("Top 5 products by quantity for each region:")
display(df_top_by_region)
```

```python
query = """
SELECT 
    DATE_TRUNC('month', order_date) as month,
    SUM(sale_price * quantity) as total_revenue,
    COUNT(*) as number_of_orders
FROM orders
WHERE EXTRACT(YEAR FROM order_date) IN (2022, 2023)
GROUP BY DATE_TRUNC('month', order_date)
ORDER BY month
"""
df_monthly = pd.read_sql(query, engine)
print("Monthly Revenue for 2022-2023:")
display(df_monthly)
```

```python
query = """
SELECT 
    EXTRACT(MONTH FROM order_date) as month,
    SUM(CASE WHEN EXTRACT(YEAR FROM order_date) = 2022 THEN sale_price * quantity END) as revenue_2022,
    SUM(CASE WHEN EXTRACT(YEAR FROM order_date) = 2023 THEN sale_price * quantity END) as revenue_2023
FROM orders
WHERE EXTRACT(YEAR FROM order_date) IN (2022, 2023)
GROUP BY EXTRACT(MONTH FROM order_date)
ORDER BY month
"""
df_comparison = pd.read_sql(query, engine)
print("Monthly Revenue Comparison 2022 vs 2023:")
display(df_comparison)
```
